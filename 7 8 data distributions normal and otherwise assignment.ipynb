```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
betavar = np.random.beta(10, 10, 100)
plt.hist(betavar)
plt.axvline(betavar.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(betavar.mean() + betavar.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(betavar.mean() - betavar.std(), color='b', linestyle='dashed', linewidth=2)
plt.show()
```

```python
binomialvar = np.random.binomial(100, .3, 100)
plt.hist(binomialvar)
plt.axvline(binomialvar.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(binomialvar.mean() + binomialvar.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(binomialvar.mean() - binomialvar.std(), color='b', linestyle='dashed', linewidth=2)
plt.show()
```

```python
chivar = np.random.chisquare(10, 100)
plt.hist(chivar)
plt.axvline(chivar.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(chivar.mean() + chivar.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(chivar.mean() - chivar.std(), color='b', linestyle='dashed', linewidth=2)
plt.show()
```

```python
dirichletvar = np.random.dirichlet((1, 2, 3), 100)
plt.hist(dirichletvar)
plt.axvline(dirichletvar.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(dirichletvar.mean() + dirichletvar.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(dirichletvar.mean() - dirichletvar.std(), color='b', linestyle='dashed', linewidth=2)
plt.show()
```

```python
expvar = np.random.exponential(10, 100)
plt.hist(expvar)
plt.axvline(expvar.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(expvar.mean() + expvar.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(expvar.mean() - expvar.std(), color='b', linestyle='dashed', linewidth=2)
plt.show()
```

```python
fvar = np.random.f(10, 10, 100)
plt.hist(fvar)
plt.axvline(fvar.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(fvar.mean() + fvar.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(fvar.mean() - fvar.std(), color='b', linestyle='dashed', linewidth=2)
plt.show()
```

Distributions chosen: beta, binomial, chisquare, dirichlet, exponential, and f
Symmetric distributions can be usefully described by basic statistics (mean and std), while the asymmetric ones rendered such values useless.

```python
norm1 = np.random.normal(5, .5, 100)
norm2 = np.random.normal(10, 1, 100)
norm3 = norm1 + norm2
plt.hist(norm3)
plt.axvline(norm3.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(norm3.mean() + norm3.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(norm3.mean() - norm3.std(), color='b', linestyle='dashed', linewidth=2)
plt.show()
n3mean = norm3.mean()
n3std = norm3.std()
print('mean: {} std: {}'.format(n3mean, n3std))
```

The mean and standard deviation of the combined distribution is the sum of the individual means and standard deviations.
